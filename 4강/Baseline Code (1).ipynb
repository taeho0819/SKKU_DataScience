## 업로드용 파일 생성 코드

```python
import pandas as pd
import numpy as np

def arr_to_csv(predictions):
    """
    predictions는 [2324.324, 342.24, 23425.34, ,,,] 처럼  charges가 담겨있는 리스트, 혹은 어레이 입니다.
    """
    df = pd.DataFrame(list(zip(list(np.arange(len(predictions))), list(predictions))), columns = ["Id","charges"])
    df.to_csv("./sumbission.csv", index=False)
```

```python
# 사용법 예시
pred = np.random.random(268)*10000 # 임의의 값을 생성해주기 위한 코드입니다.

# pred대신의 여러분의 예측값을 넣어서 csv파일을 뽑아야 합니다.
arr_to_csv(pred)
```
