## CSV로 내보내기.

```python
import numpy as np
import pandas as pd
```

```python
def arr_to_csv(predictions):
    """
    predictions는 [3,2,4,3,5,4,4,,, ] 처럼 와인의 퀄리티가 담겨있는 리스트, 혹은 어레이 입니다.
    """
    df = pd.DataFrame(list(zip(list(np.arange(len(predictions))), list(predictions))), columns = ["Id","Predicted"])
    df.to_csv("./sumbission.csv", index=False)
```

```python
pred = (np.random.random([299,])*10).astype(int)
pred
```

```python
# 사용법
arr_to_csv(pred)
```
